```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

配列の作成

```python
digit = 9
# 10^9個の列を作成
list = [0] * 10**digit
# データを格納する配列
data = []
# 1000個のデータの平均を取り，ヒストグラムに加える，という操作を10000回繰り返す
for i in range (0,10000):
	x_mean = 0
	N = 1000
	for n in range(0, N-1):
		# 0〜1の
		a = random.random()
		x_mean += a
	x_mean /= N
	num = int(x_mean*10**digit)
	list[num] += 1
	data.append(num)
plt.hist(data,bins=100)
```
